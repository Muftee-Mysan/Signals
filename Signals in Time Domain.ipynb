#### Example 2. The following example illustrates plotting two waves in the same plot
Plot of ùë•(ùë°) = ùê¥cos(ùúî0ùë° + ùúô) with ùúô = 0 and ùúô = ùúã/2

```python
fs = 4.4e4 #sampling frequency
ts = 1 / fs
t = np.arange(-1.,1.,ts)
f = 2 # 2 Hz
w0 = 2*np.pi*f
phi = 0
x1t = np.cos(w0*t+phi)
phi = np.pi / 2
x2t = np.cos(w0*t + phi)
fig, ax = plt.subplots()
ax.plot(t,x1t, label='$\cos(4\pi t)$', color = 'r')
ax.plot(t,x2t, label='$\cos(4\pi t + \pi/2)$', color = 'b')
ax.set_xlabel('$t$')
ax.set_ylabel('$x(t)$')
plt.legend(loc = 'lower right')
ax.grid(True)

plt.show()
```

## 1.2 Real CT Exponential


```python
fs = 4.4e4 # sampling frequency 
ts = 1 / fs
t = np.arange(-1.,1.,ts)

#Your code goes here
C = 1
alpha = 1.2
xt = C * np.exp(alpha*t)
fig,ax = plt.subplots()
ax.plot(t,xt)
ax.set_xlabel('$t$')
ax.set_ylabel('$x(t)$')
ax.grid(True)

plt.show()
```

#### Comment on the effect due to the sign of ùõº on the signal.

If the sign of alpha is negative, the graph is a decreasing exponential, tending to 0 with infinity.
If the sign of alpha is positive, the signal is an increasing exponential, tending to infinity. 


```python
fs = 4.4e4 # sampling frequency 
ts = 1 / fs
t = np.arange(-1.,1.,ts)

#Your code goes here
C = 1
alpha = -1.2
xt = C * np.exp(alpha*t)
fig,ax = plt.subplots()
ax.plot(t,xt)
ax.set_xlabel('$t$')
ax.set_ylabel('$x(t)$')
ax.grid(True)

plt.show()
```

 ## 1.3 Growing Sinusoidal Signal 

```python

fs = 4.4e4  # sampling frequency
ts = 1 / fs
t = np.arange(-1., 1., ts)
f = 5  # 5 Hz

w0 = 2 * np.pi * f

xt1 = 0.15 * np.exp(2 * t) * np.cos(w0 * t + 0)
xt2 = 0.15 * np.exp(-2 * t) * np.cos(w0 * t + np.pi / 2)

fig, axes = plt.subplots(2, 1, sharex='all', sharey='all', figsize=(18, 18))

axes[0].plot(t, xt1, color='y')
axes[0].set_xlabel('$t$')
axes[0].set_ylabel('$xt1$')
axes[0].grid(True)
axes[0].title.set_text('$0.15e^{2t}\cos(w0t)$')

axes[1].plot(t, xt2, color='b')
axes[1].set_xlabel('$t$')
axes[1].set_ylabel('$xt2$')
axes[1].grid(True)
axes[1].title.set_text('$0.15e^{-2t}\cos(w0t + \pi/2)$')

plt.show()
```

## 1.7 Transformation of the Independent Variable (Graded)

```python
 def x(t):
     if (t < 0.):
         return 0.
     elif (t < 1.):
         return 1.
     elif (t < 2.):
         return 2.- t
     else:
         return 0.
 
 fs = 4.4e4 # 44,000‚àíHz sampling frequency
 ts = 1/fs
 t = np.arange(-3.5, 3.5, ts) # A linearly‚àíspaced array with step ts
 fig, axes = plt.subplots(7,1, sharey='all', figsize=(10,15))
 
 # x(t)
 axes[0].plot(t, [x(t_) for t_ in t], color = 'r')
 axes[0].set_xlabel('$t$')
 axes[0].set_ylabel('$x(t)$')
 axes[0].grid(True)
 
 axes[1].plot(t, [x(t_) for t_ in (t-1)], color = 'b')
 axes[1].set_xlabel('$t$')
 axes[1].set_ylabel('$x(t-1)$')
 axes[1].grid(True)
    
 axes[2].plot(t, [x(t_) for t_ in (t+1)], color = 'y')
 axes[2].set_xlabel('$t$')
 axes[2].set_ylabel('$x(t+1)$')
 axes[2].grid(True)
    
 axes[3].plot(t, [x(t_) for t_ in (-t)], color = 'g')
 axes[3].set_xlabel('$t$')
 axes[3].set_ylabel('$x(-t)$')
 axes[3].grid(True)
    
 axes[4].plot(t, [x(t_) for t_ in (-t+1)], color = 'purple')
 axes[4].set_xlabel('$t$')
 axes[4].set_ylabel('$x(-t+1)$')
 axes[4].grid(True)
    
 axes[5].plot(t, [x(t_) for t_ in (3*t/2)], color = 'black')
 axes[5].set_xlabel('$t$')
 axes[5].set_ylabel('$x(3t/2)$')
 axes[5].grid(True)
    
 axes[6].plot(t, [x(t_) for t_ in (((3*t)/2) +1)], color = 'pink')
 axes[6].set_xlabel('$t$')
 axes[6].set_ylabel('$x(3t/2 + 1)$')
 axes[6].grid(True)

    
 # Your code goes here
 plt.show()
```

## 1.8 Observing a Signal in Frequency Domain

The following code plots a signal and its frequency domain representation. Note that the non-zero values
 are actually delta functions in frequency, which we will learn later.

```python
 fs = 100# 100‚àíHz sampling frequency
 ts = 1/fs
 t = np.arange(-1., 1., ts) # A linearly‚àíspaced array with step ts
 fig, axes = plt.subplots(2,1, figsize=(10,10))

 f = 2#2Hz
 omega0 = 2*np.pi*f

 xt = 0.75 + 2.*np.cos(omega0*t) + 1.*np.cos(3*omega0*t)
    
 Xf = np.fft.fft(xt)
 freq = np.fft.fftfreq(t.shape[-1], d=ts)
    
 axes[0].plot(t,xt)
 axes[0].set_xlabel('$t$ in s')
 axes[0].set_ylabel('$x(t)$')
 valsubrange = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))
 freqsubrage = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))
 axes[1].stem(freq[freqsubrage], Xf.real[valsubrange]/len(t))
 axes[1].set_xlabel('$f$ in Hz')
 axes[1].set_ylabel('$\mathfrak{Re}(X(j\omega))$')
 plt.xticks(np.arange(-10,11))
 plt.show()
```

#### Interpret the frequency domain representation. 
 

The frequency domain of the signal consists of the two main frequencies of 2Hz and 6Hz. They have amplitudes of 1.0 and 0.5 respectively. Also, because of the DC shift of 0.75, we also get a spike in 0 Hz which has an amplitude of 0.75.
